```python
import tensorflow as tf
import numpy
import matplotlib.pyplot as plt
rng = numpy.random
optimizer=[]
# Parameters
learning_rate = [0.005, 0.0005]
training_epochs = [200,500,1000]
display_step = 50

# Training Data
train_X = numpy.asarray([1,2,4,3,5])
train_Y = numpy.asarray([1,3,3,2,5])
n_samples = train_X.shape[0]

# tf Graph Input
X = tf.placeholder("float")
Y = tf.placeholder("float")

# Set model weights
W = tf.Variable(0.0, name="weight")
b = tf.Variable(0.0, name="bias")

# Construct a linear model
pred = tf.add(tf.multiply(X, W), b)


# Mean squared error
cost = tf.reduce_sum(tf.pow(pred-Y, 2))/(2*n_samples)
# Gradient descent
optimizer.append(tf.train.GradientDescentOptimizer(learning_rate[0]).minimize(cost))
optimizer.append(tf.train.GradientDescentOptimizer(learning_rate[1]).minimize(cost))

# Initializing the variables
init = tf.global_variables_initializer()

# Launch the graph
with tf.Session() as sess:
    sess.run(init)
    for ab in range(2):
        print("\nFOR   Learning Rate\t",learning_rate[ab],"\n")
        for training_epo in training_epochs:
            print("\nFOR \t",training_epo,"\n")
        # Fit all training data
            for epoch in range(training_epo):
                for (x, y) in zip(train_X, train_Y):
                    sess.run(optimizer[ab], feed_dict={X: x, Y: y})

                #Display logs per epoch step
                if (epoch+1) % display_step == 0:
                    c = sess.run(cost, feed_dict={X: train_X, Y:train_Y})
                    print ("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(c), \
                        "W=", sess.run(W), "b=", sess.run(b))

            print ("Optimization Finished!")
            training_cost = sess.run(cost, feed_dict={X: train_X, Y: train_Y})
            print ("Training cost=", training_cost, "W=", sess.run(W), "b=", sess.run(b), '\n')

            #Graphic display
            plt.plot(train_X, train_Y, 'ro', label='Original data')
            plt.plot(train_X, sess.run(W) * train_X + sess.run(b), label='Fitted line')
            plt.legend()
            plt.show()
            print("\n*********************************************************************************************\n")
```
